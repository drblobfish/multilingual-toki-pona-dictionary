```python
import bs4 as bs
import os
import os.path as path
import pandas as pd
```

# parsing

```python
def parse():
    langs = [i for i in os.listdir("data") if len(i)==2]
    
    dico_setup = False
    
    for lang in langs:
    
        with open(path.join("data",lang,"index.html")) as fp:
            soup = bs.BeautifulSoup(fp, 'html.parser')
        
        dico = soup.find_all(class_="dictionary")
        assert len(dico)==1 ,"multiple dictionary find in the page"
        dico = dico[0]
        
        nimiH5 = [i.text for i in dico.find_all("h5")]
        trans = [i.text for i in dico.find_all("p")]
        assert len(nimiH5)==120 , "not the correct amount of nimis"
        assert len(trans)==120 , "not the correct amount of translation"
        
        if not dico_setup:
            df = pd.DataFrame(index=nimiH5,columns=langs)
            dico_setup = True
        
        for nim,tran in zip(nimiH5,trans):
            df.loc[nim,lang] = tran
        
    return df
```

```python
df = parse()
```
